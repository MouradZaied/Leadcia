```
# Importing dependencies
import matplotlib.pyplot as plt
import numpy as np
from scipy import misc
```

```
def function(x):
    return np.sinc(0.48*x)
```

```
x = np.arange(-15, 15, 0.01)
y = function(x)

fig = plt.figure(figsize=(14,8))
plt.plot(x, y,'r-')
plt.grid()
plt.show()
```

```
alpha = 0.2 # learning rate
nb_max_iter = 100 # Nb max d'iteration
eps = 0.0001 # stop condition
#x=0.9 # intinial value of x
x=-5.6 # intinial value of x
```

```
y=function(x)
x_cache=[]
y_cache=[]
x_cache.append(x)
y_cache.append(y)

cond = eps + 10.0 # start with cond greater than eps (assumption)
nb_iter = 0 
tmp_y = y
while cond > eps and nb_iter < nb_max_iter:
  x = x - alpha * misc.derivative(function, x)
  y = function(x)
  nb_iter = nb_iter + 1
  cond = abs( tmp_y - y )
  tmp_y = y
  x_cache.append(x)
  y_cache.append(y)
```

```
x = np.arange(-15.0, 15.0, 0.01)
y = function(x)

fig = plt.figure(figsize=(14,8))
plt.plot(x, y,'r-')
plt.plot(x_cache, y_cache,'bo')
plt.grid()
plt.show()
```

```
x = np.arange(-15.0, 15.0, 0.01)
y = function(x)
fig = plt.figure(figsize=(14,8))
ax = fig.add_subplot(111)
ax.grid()
ax.plot(x, y,'r-')
plt.show()
```

```
plt1, = ax.plot([], [], 'bo', label='Initial hypothesis')
text_box = ax.text(-3.7, 70, 'Iteration 0', size = 16,color='green')
```

```
def init():
    plt1.set_data([], [], 'bo')
    text_box.set_data(-3.7, 70, 'Iteration 0', size = 16,color='green')
    return (plt1, text_box)
```

```
def animate(i):
    x = x_cache[0:i]
    y = y_cache[0:i]
    plt1.set_data(x, y)
    text_box.set_text('Iteration: {}'.format(i))
    # plt.legend()
    return (plt1, text_box)
```

```
anim = animation.FuncAnimation(fig, animate, frames=35, interval=1000, blit=True)
```
