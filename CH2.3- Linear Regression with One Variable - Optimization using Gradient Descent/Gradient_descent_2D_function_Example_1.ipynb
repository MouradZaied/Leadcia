```python
# importing dependencies
import matplotlib.pyplot as plt
import numpy as np
from scipy import misc
```

```python
def function(x):
    return 3*x**2+2*x+1
```

```python
x = np.arange(-2,2,0.01)
y = function(x)

fig = plt.figure(figsize=(14,8))
plt.plot(x,y,'r-')
plt.grid()
plt.show()
```

```python
alpha = 0.02  # Learning rate
nb_max_iter = 100 # Nb max iterations
eps = 0.0001 # stop condition error
x = 2 # Initial value of x
```

```python
# gradient descent algorithm
y = function(x)
x_cache=[]
y_cache=[]
x_cache.append(x)
y_cache.append(y)

cond = eps + 10.0 # start with greater than eps
nb_iter = 0
tmp_y = y

while cond > eps and nb_iter < nb_max_iter :
  x = x - alpha * misc.derivative (function, x)
  y = function(x)
  nb_iter = nb_iter + 1
  cond = abs (tmp_y - y)
  tmp_y = y
  x_cache.append(x)
  y_cache.append(y)
```

```python
x = np.arange(-2.0, 2.0, 0.01)
y = function(x)


fig = plt.figure(figsize=(14,8))
plt.plot(x, y,'r-')
plt.plot(x_cache, y_cache,'bo')
plt.grid()
plt.show()
```

Animate the gradient descent algorithm

```python
from matplotlib import animation, rc
from IPython.display import HTML
rc('animation', html='html5')
```

```python
x = np.arange(-2.0, 2.0, 0.01)
y = function(x)


fig = plt.figure(figsize=(14,8))
ax=fig.add_subplot(111)
plt.plot(x, y,'r-')
plt.grid()
plt.show()
```

```python
plt1, = ax.plot([], [], 'bo', label='Initial hypothesis')
text_box = ax.text(-2, 15, 'Iteration 0', size = 16,color='green')
```

```python
def init():
    plt1.set_data([],[],'bo')
    text_box = ax.text(-2, 15, 'Iteration 0', size = 16, color='green')
    return (plt1, text_box)
```

```python
def animate(i):
    x = x_cache[0:i]
    y = y_cache[0:i]
    plt1.set_data(x, y)
    text_box.set_text('Iteration: {}'.format(i))
    return (plt1, text_box)
```

```python
anim = animation.FuncAnimation(fig, animate, frames=20, interval=1000, blit=True)
```
